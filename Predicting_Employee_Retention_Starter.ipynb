2.1.3 Handle rows with missing values <font color="red">[4 Marks]</font>

## The gold standard here is multiple imputation (MI) (aka â€œMICEâ€ or chainedâ€‘equations). MIâ€¦

1. Preserves all your rows,

2. Draws from the posterior to reflect imputation uncertainty (so your standard errors stay honest), and

3. Under a Missingâ€‘Atâ€‘Random assumption, introduces less bias than singleâ€‘value imputation.

```python
# Handle the missing value rows in the column
# columns to impute
import pandas as pd

from sklearn.experimental import enable_iterative_imputer   # noqa
from sklearn.impute import IterativeImputer
from sklearn.linear_model import BayesianRidge
cols = ['Distance from Home', 'Company Tenure (In Months)']

# 1) set up the MI engine
mi = IterativeImputer(
    estimator=BayesianRidge(),
    sample_posterior=True,     # draws from the predictive posterior
    max_iter=10,               # number of MICE cycles
    random_state=42
)

# 2) fit & transform
df_imputed = df.copy()
df_imputed[cols] = mi.fit_transform(df[cols])
```

2.1.4 Check percentage of remaning data after missing values are removed <font color="red">[2 Mark]</font>

```python
# Check the percentage of remaining data after missing values are removed
n_original = len(df)
n_missing_after = df_imputed[cols].isna().any(axis=1).sum() 
n_after = len(df_imputed)
pct_retained = n_after / n_original * 100
print(f"Rows before handling missing data: {n_original}")
print(f"Rows after handling missing data:  {n_after}")
print(f"Percentage of data retained:       {pct_retained:.2f}%")
```

7.3.1 Plot ROC curve <font color="red">[3 Marks]</font>

```python
# 7.3.1 Define and plot the ROC curve (with AUC)

import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, roc_auc_score

def plot_roc_curve(y_true, y_prob):
    """
    Plots ROC curve and returns
    fpr, tpr, thresholds.
    """
    # Compute false positive rate, true positive rate, thresholds
    fpr, tpr, thresholds = roc_curve(y_true, y_prob)
    # Compute area under the curve
    auc_score = roc_auc_score(y_true, y_prob)

    # Plot
    plt.figure(figsize=(6, 6))
    plt.plot(fpr, tpr, label=f"AUC = {auc_score:.3f}")
    plt.plot([0, 1], [0, 1], "--", color="gray", label="Chance")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("ROC Curve")
    plt.legend(loc="lower right")
    plt.tight_layout()
    plt.show()

    return fpr, tpr, thresholds
```

```python
# Create confusion matrix
# 8.3 Create confusion matrix on the validation set
from sklearn.metrics import confusion_matrix

cm_val = confusion_matrix(y_val_bin, df_val_results['final_prediction'])
print("Validation Confusion Matrix:\n", cm_val)

# 8.3 â€“ Extract TP, TN, FP, FN
tn_val, fp_val, fn_val, tp_val = cm_val.ravel()

print(f"True Negatives : {tn_val}")
print(f"False Positives: {fp_val}")
print(f"False Negatives: {fn_val}")
print(f"True Positives : {tp_val}")
```

### **8.4 Calculate sensitivity and specificity** <font color = red>[4 marks]</font>

```python
# 8.4 Calculate sensitivity and specificity on the validation set
# ---------------------------------------------------------------
# We already extracted tn_val, fp_val, fn_val, tp_val from cm_val

# Sensitivity (Recall of â€œLeftâ€ class)
sensitivity_val = tp_val / (tp_val + fn_val)

# Specificity (Recall of â€œStayedâ€ class)
specificity_val = tn_val / (tn_val + fp_val)

print(f"Sensitivity (Recall â€“ Left): {sensitivity_val:.3f}")
print(f"Specificity (Stayed):        {specificity_val:.3f}")
```

### **8.5 Calculate precision and recall** <font color = red>[4 marks]</font>

```python
# 8.5 Calculate precision and recall of the model on the validation set
# --------------------------------------------------------------------
from sklearn.metrics import precision_score, recall_score

precision_val = precision_score(y_val_bin, df_val_results['final_prediction'])
recall_val    = recall_score(y_val_bin, df_val_results['final_prediction'])   # same as sensitivity_val

print(f"Precision (Left): {precision_val:.3f}")
print(f"Recall    (Left): {recall_val:.3f}")
```
